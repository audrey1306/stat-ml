
## Exercice 2 : maximum de vraisemblance

### <span style="color:blue">QUESTION 2.1 :</span> 

Tirer 10 fois une pièce à pile ou face et modéliser les résultats obtenus comme ceux d'une variable aléatoire X qui vaut X_i=0 si on a pile et X_i=1 si on a face.

Calculez le maximum de vraisemblance du paramètre p d'un loi de Bernoulli qui modéliserait le problème. Pour y arriver, différentes valeures possibles de p seront testées et le p retenu sera celui qui a la plus grande vraisemblance.


- Vérifier empiriquement comment évolue ce maximum de vraisemblance si l'on effectue de plus en plus de tirages
- Que se passe-t-il quand il y a trop de tirages ? Représenter la log-vraisemblance plutot que la vraisemblance dans ce cas.




```python
NbTirages=500
NbPiles = np.random.randn(NbTirages) > 0
NbPiles = np.sum(NbPiles)
NbFaces=NbTirages-NbPiles


PossibleValuesForP=np.linspace(0.01,0.99,1000)



def vraisemblance(n_pile, n_face, p):
    return p**n_pile * (1-p)**n_face


def log_vraisemblance(n_pile, n_face, p):
    return n_pile * np.log(p) + n_face * np.log(1 - p) #si np.log(p**n_pile * (1-p)**n_face) : risque d'avoir log(0)



tab_vraisemblance = [vraisemblance(NbPiles, NbFaces,p) for p in PossibleValuesForP]
argmax_tab_vraisemblance = np.argmax(tab_vraisemblance)
p_max = PossibleValuesForP[argmax_tab_vraisemblance]

print("p_max : ", p_max)

tab_logvraisemblance = [log_vraisemblance(NbPiles, NbFaces,p) for p in PossibleValuesForP]
argmax_tab_logvraisemblance = np.argmax(tab_logvraisemblance)
p_max_log = PossibleValuesForP[argmax_tab_logvraisemblance]

print("p_max log : ", p_max_log)


plt.plot(PossibleValuesForP, tab_vraisemblance)
```

Vérifier empiriquement comment évolue ce maximum de vraisemblance si l'on effectue de plus en plus de tirages.
```python
NbTirages=np.linspace(1,1000,1000, dtype=int)
P = np.zeros(len(NbTirages))




for n in range(len(NbTirages)):
    NbPiles = np.random.randn(NbTirages[n]) > 0
    NbPiles = np.sum(NbPiles)
    NbFaces=NbTirages[n]-NbPiles

    PossibleValuesForP=np.linspace(0.01,0.99,100)

    tab_logvraisemblance = [log_vraisemblance(NbPiles, NbFaces,p) for p in PossibleValuesForP]
    argmax_tab_logvraisemblance = np.argmax(tab_logvraisemblance)
    P[n] = PossibleValuesForP[argmax_tab_logvraisemblance]



plt.plot(NbTirages, P)
```
